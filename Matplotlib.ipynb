#MATPLOTLIB CONCEITOS BASICOS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,30)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
plt.plot(x,y)
#para definir os eixos, sempre a primeira opção vai no x e a segunda no y
# no colab a função plt.show() não se faz necessaria, mas em outros compuladores ela se torna essencial para
# mostrar o grafico
# plt.show()
```

```python
plt.plot(x,y, "g-*")
plt.xlabel("Valores Eixo X")
plt.ylabel("Valores Eixo Y")
plt.title("Grafico X vs Y")
```

```python
plt.bar(x,y)
plt.xlabel("Valores Eixo X")
plt.ylabel("Valores Eixo Y")
plt.title("Grafico X vs Y")
```

```python
# usamos a função subplot para colocar mais de um grafico em um bloco de cogido
# nela passamos 3 argumentos, o primeiro é a quantidade de linhas o segundo é a quantidade de colunas
# a terceira é a posição do grafico
```

```python
plt.subplot(1,2,1)
plt.subplot(1,2,2)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, "g-*")
plt.xlabel("Valores Eixo X")
plt.ylabel("Valores Eixo Y")
plt.title("Grafico X vs Y")

plt.subplot(1,2,2)
plt.plot(x,y**2, "r-*")
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y, "r--")
plt.subplot(1,2,2)
plt.plot(x,y**2)
```

#PARTE 2 MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,30)
y = x **2
```

```python
from inspect import getframeinfo
fig, axes = plt.subplots(3,2)
# diferente do subplot que cria os graficos conforme informados, 1 por 1, o metodo subplots cria um "array" de graficos
# sempre na ordem linha | colunas
# depois o acesso passamos com a variavel que foi salvo, no caso ax e passando o parametro de qual grafico quero acessar como por exemplo
axes[1,1].plot(x,y)
#onde ele alterar somente o grafico necessario
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,5))
ax[0,0].plot(x,y,"r--")
ax[0,0].set_xlabel("Eixo x primeiro grafico") #incluindo o titulo eixo x no grafico 0,0
ax[0,0].set_ylabel("Eixo y primeiro grafico") #incluindo o titulo eixo y no grafico 0,0
ax[0,0].set_title("Primeiro grafico") #incluindo o titulo no grafico 0,0
ax[0,1].plot(x,y**2,"b--")
ax[1,0].plot(x,y**3,"g--")
ax[1,1].bar(x,y**4)

plt.tight_layout() #função para tentar de forma automatica ajustar o layout dos graficos
```

#USANDO PANDAS + MATPLOTLIB

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salaries.csv")
```

```python
df
```

```python
df = df[df["BasePay"] != "Not Provided"]
```

```python
df["BasePay"] = df["BasePay"].astype(float)
```

```python
df_plots = df [["BasePay","JobTitle","Year"]]
```

```python
df_plots
```

```python
df_plots.groupby("Year").size()
```

```python
df_plots["Year"] = df["Year"].astype(str)
qtd_por_ano = df_plots.groupby("Year").size()
```

```python
plt.figure(figsize=(8,5))
plt.bar(qtd_por_ano.index, qtd_por_ano.values, color="g")
plt.xlabel("Ano")
plt.ylabel("Quantidade de pessoas")
plt.title("Quantidade por ano")
```

```python
df["Year"] = df["Year"].astype(str)
media_salario = df.groupby("Year")["BasePay"].mean()
```

```python
plt.figure(figsize=(8,5))
plt.plot(media_salario.index, media_salario.values, "g--")
plt.xlabel("Ano")
plt.ylabel("Media Salarial")
plt.title("Media Salarial por ano")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax[0].bar(qtd_por_ano.index, qtd_por_ano.values, color="green")
ax[0].set_xlabel("Ano")
ax[0].set_ylabel("Quantidade de pessoas")
ax[0].set_title("Quantidade por ano")

ax[1].plot(media_salario.index, media_salario.values, "g--")
ax[1].set_xlabel("Ano")
ax[1].set_ylabel("Media Salarial")
ax[1].set_title("Media Salarial por ano")
plt.tight_layout()
```
